```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

min_mse = float('inf')
best_rs = 0

for i in range(1, 1000):
    x_train, x_test, y_train, y_test = train_test_split(xav, yav, test_size=0.20, random_state=i)
    lr = LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, pred)
    r2 = r2_score(y_test, pred)
    
    if mse < min_mse:
        min_mse = mse
        best_rs = i
print('Best MSE:', min_mse, 'at random_state', best_rs)
print("MSE:", mse,'at random_state', best_rs)
print("RMSE:", rmse,'at random_state', best_rs)
print("MAE:", mae),'at random_state', best_rs
print("R-squared:", r2,'at random_state', best_rs)
```
